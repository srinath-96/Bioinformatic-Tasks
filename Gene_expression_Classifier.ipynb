```python
# Convert to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train.values)
X_test_tensor = torch.FloatTensor(X_test.values)
y_train_tensor = torch.LongTensor(y_train.values)
y_test_tensor = torch.LongTensor(y_test.values)

print(f"Training set: {X_train_tensor.shape}")
print(f"Test set: {X_test_tensor.shape}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class GeneExpressionNet(nn.Module):
    def __init__(self, input_size=2, hidden_size=16, num_classes=2):
        super(GeneExpressionNet, self).__init__()
        
        # Define layers
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.2)
        
        self.fc2 = nn.Linear(hidden_size, hidden_size//2)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.2)
        
        self.fc3 = nn.Linear(hidden_size//2, num_classes)
        
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.dropout1(x)
        
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.dropout1(x)
        
        x = self.fc3(x)
        return x

# Initialize the model
model = GeneExpressionNet(input_size=2, hidden_size=32, num_classes=2)
print(model)

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

print(f"\nModel parameters: {sum(p.numel() for p in model.parameters())}")
```

```python
# Training function
def train_model(model, X_train, y_train, X_test, y_test, epochs=200):
    from sklearn.metrics import f1_score, classification_report, accuracy_score
    train_losses = []
    test_f1_scores = []
    
    for epoch in range(epochs):
        # Training
        model.train()
        optimizer.zero_grad()
        
        # Forward pass
        outputs = model(X_train)
        loss = criterion(outputs, y_train)
        
        # Backward pass
        loss.backward()
        optimizer.step()
        
        train_losses.append(loss.item())
        
        # Evaluation every 20 epochs
        if (epoch + 1) % 20 == 0:
            model.eval()
            with torch.no_grad():
                test_outputs = model(X_test)
                _, predicted = torch.max(test_outputs.data, 1)
                
                f1 = f1_score(y_test, predicted.numpy(), average='macro')
                test_f1_scores.append(f1)
                
                print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}, Test F1 Macro: {f1:.4f}')
    
    return train_losses, test_f1_scores

# Train the model
print("Training Neural Network...")
train_losses, test_f1_scores = train_model(model, X_train_tensor, y_train_tensor, 
                                          X_test_tensor, y_test_tensor, epochs=200)
```

```python
# Final evaluation
model.eval()
with torch.no_grad():
    import matplotlib.pyplot as plt
    from sklearn.metrics import f1_score, classification_report, accuracy_score
    # Training predictions
    train_outputs = model(X_train_tensor)
    _, train_predicted = torch.max(train_outputs.data, 1)
    
    # Test predictions
    test_outputs = model(X_test_tensor)
    _, test_predicted = torch.max(test_outputs.data, 1)
    
    # Calculate metrics
    train_f1 = f1_score(y_train, train_predicted.numpy(), average='macro')
    test_f1 = f1_score(y_test, test_predicted.numpy(), average='macro')
    
    train_acc = accuracy_score(y_train, train_predicted.numpy())
    test_acc = accuracy_score(y_test, test_predicted.numpy())
    
    print(f"\n=== Final Results ===")
    print(f"Training F1 Macro: {train_f1:.4f}")
    print(f"Test F1 Macro: {test_f1:.4f}")
    print(f"Training Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")
    
    print(f"\nDetailed Classification Report:")
    print(classification_report(y_test, test_predicted.numpy()))

# Plot training progress
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(train_losses)
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.subplot(1, 2, 2)
epochs_f1 = range(19, len(train_losses), 20)
plt.plot(epochs_f1, test_f1_scores, 'r-o')
plt.title('Test F1 Macro Score')
plt.xlabel('Epoch')
plt.ylabel('F1 Macro')

plt.tight_layout()
plt.show()
```

```python
def predict_new_sample(model, gene_one_val, gene_two_val):
    """
    Predict class for new gene expression values
    """
    import numpy as np
    model.eval()
    
    # Prepare input
    new_sample = np.array([[gene_one_val, gene_two_val]])
    new_sample_tensor = torch.FloatTensor(new_sample)
    
    # Make prediction
    with torch.no_grad():
        output = model(new_sample_tensor)
        probabilities = torch.softmax(output, dim=1)
        _, predicted = torch.max(output.data, 1)
        
    return predicted.item(), probabilities.numpy()[0]

# Example usage
pred_class, pred_probs = predict_new_sample(model, 5.0, 4.0)
print(f"\nPrediction for Gene One=5.0, Gene Two=4.0:")
print(f"Predicted Class: {pred_class}")
print(f"Class Probabilities: [Class 0: {pred_probs[0]:.3f}, Class 1: {pred_probs[1]:.3f}]")
```
